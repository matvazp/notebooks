The goal is to fabricate a stack of __hBN/Gr/hBN/Gr/hBN__ supported on a silicon nitride membrane with a small aperture. Across the stack runs a 20nm-width nanopore that serves as the channel where separations are carried out by the influence of an electric field. A flow cell is assembled around the device to provide mechanical support and protection, as well as electrical and fluidic connectivity. Electrically, the flow cell allows connection from the graphene monolayers in the stack to external tools (waveform generator, oscilloscope, etc.). Fluidically, the flow cell connects a reservour to each side of the stack, allowing only the fluid to interact across the nanopore.

## Device Fabrication

For the top layer, an hBN flake on silicon dioxide substrate is spin-coated with PPA. The PPA is patterned with the Nanofrazor as a soft mask, and exposed to SF6 plasma to etch both the PPA and hBN, so the pattern is transfered to the hBN flake. PPA is then removed with a post-etching bake to expose the patterned hBN with a nanohole of around 20 nm. 

For the bottom layer, the hBN flake is suspended on a silicon nitride membrane previously patterned using a focused ion beam, so the flake can be etched buttom up. The aperture on the membrane works as a hard mask to allow the plasma to reach only the exposed area of the hBN flake.

The Graphene/hBN/Graphene stack is transfered on top of the bottom hBN ensuring the graphene layers contact the gold pads on the membrane. Then the top hBN follows to complete the stack. Graphene layers are etched with oxygen plasma, and a final SF6 plasma step etches the middle hBN layer to create the channel that runs across the stack.

```python
# Create a single figure
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax1.imshow(img)
ax1.axis('off')  # Turn off the axis numbers and ticks
ax1.set_title('AFM image of a FIB-defined nanohole')

# Profile to the right
ax2 = fig.add_subplot(122)
df.plot(ax=ax2, x='x', y='y', stacked=True)
ax2.set_title('Profile of hole with TAP150A probe')
ax2.set_xlabel('Distance (m)')
ax2.set_ylabel('Height sensor (m)')
ax2.axvspan(3.2e-7, 5.2e-7, color='red', alpha=0.2, visible=True, label="sa")
ax2.text(4.2e-7, 1.4e-8, "200 nm", ha="center", va="center", fontsize=10)
plt.annotate('', xy=(3.2e-7, 1.3e-8), xycoords='data', xytext=(5.2e-7, 1.3e-8), 
             textcoords='data', arrowprops={'arrowstyle': '<->'})

# Display the combined figure
plt.tight_layout()
plt.show()
```
