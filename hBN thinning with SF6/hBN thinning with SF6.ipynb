# SF6 etching (17/October/2023)
We put exfoliated hBN on a SiO2 chip in the plasma chamber to use it as a reference for the backside etching tests. 

|Parameter|Test 1 | Test 2|
|---|---|---|
|Plasma?         | No | **Yes** |
|Gas flow (sccm) |  5 |  4 |
|Power (W)       | 25 | 20 |
|Time (seconds)  | 70 | 60 |
|Bias (V)        |  2 | 30 |
|Pressure (Pa)   | 7.97E-3 | 8.25E-3 |

# AFM analysis (25/October/2023)

With the AFM analysis we can conclude that after 60 seconds of SF6 plasma on the surface of the device, hBN was thinned down to thicknesses below 10 nm. However, __severe contamination__ is noticed on both, the hBN flakes and the SiO2 substrate in the form of small dots that are quite similar to those observed in the AFM analysis of the backside etched hBN flakes. 

```python
img_f1l1 = plt.imread('data/flake01_optical_01.tif')
img_f3l1 = plt.imread('data/flake03_optical_01.tif')
img_f4l1 = plt.imread('data/flake04_optical_01.tif')
img_f5l1 = plt.imread('data/flake05_optical_01.tif')

afm_f1l1 = plt.imread('data/SF6hBN_Flake1_Location1_CORRECTED.jpg')
afm_f3l1 = plt.imread('data/SF6hBN_Flake3_Location1_CORRECTED.jpg')
afm_f4l1 = plt.imread('data/SF6hBN_Flake4_Location1_CORRECTED.jpg')
afm_f5l1 = plt.imread('data/SF6hBN_Flake5_Location2_CORRECTED.jpg')
```

```python
afm01 = plt.figure(figsize=(10,10))
ax1 = afm01.add_subplot(2, 2, 1)  # 1 row, 2 columns, first plot
ax1.text(150, 250, "5.21 nm", ha="center", va="center", fontsize=10)
ax1.axis('off')  # Turn off the axis numbers and ticks
ax1.set_title('FLAKE 1')
ax1.imshow(img_f1l1)

ax1 = afm01.add_subplot(2, 2, 2)  # 1 row, 2 columns, first plot
ax1.text(150, 250, "3.011 nm", ha="center", va="center", fontsize=10)
ax1.axis('off')  # Turn off the axis numbers and ticks
ax1.set_title('FLAKE 2')
ax1.imshow(img_f3l1)

ax1 = afm01.add_subplot(2, 2, 3)  # 1 row, 2 columns, first plot
ax1.text(200,90, "2.623 nm", ha="center", va="center", fontsize=10)
plt.annotate('', xy=(170, 120), xycoords='data', xytext=(200,100), 
             textcoords='data', arrowprops={'arrowstyle': '->'})
ax1.axis('off')  # Turn off the axis numbers and ticks
ax1.set_title('FLAKE 3')
ax1.imshow(img_f4l1)

ax1 = afm01.add_subplot(2, 2, 4)  # 1 row, 2 columns, first plot
ax1.text(200, 90, "2.186 nm", ha="center", va="center", fontsize=10)
plt.annotate('', xy=(160, 120), xycoords='data', xytext=(200,100), 
             textcoords='data', arrowprops={'arrowstyle': '->'})
ax1.text(200, 180, "6.798 nm", ha="center", va="center", fontsize=10)
plt.annotate('', xy=(180, 140), xycoords='data', xytext=(200,170), 
             textcoords='data', arrowprops={'arrowstyle': '->'})
ax1.axis('off')  # Turn off the axis numbers and ticks
ax1.set_title('FLAKE 4')
ax1.imshow(img_f5l1)

plt.show()
```
